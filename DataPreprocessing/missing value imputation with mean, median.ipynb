```python
#function
new2=[var for var in New2.columns if New2[var].isnull().sum()>0 ]
```

```python
new2
```

```python
for i, var in enumerate(new2):
    plt.subplot(2,2,i+1)
    sns.distplot(New2[var],bins=20)
```

```python
New3=New2.fillna(New2.mean())
New3.isnull().sum().sum()
```

```python
plt.figure(figsize=(16,9))

for i, var in enumerate(new2):
    plt.subplot(2,2,i+1)
    
    sns.distplot(df[var],bins=20,label='original dataset')
    sns.distplot(New3[var],bins=20,label='cleaned by mean dataset')
    plt.legend()
```

```python
New4=New2.fillna(New2.median())
New4.isnull().sum().sum()
```

```python
plt.figure(figsize=(16,9))

for i, var in enumerate(new2):
    plt.subplot(2,2,i+1)
    
    sns.distplot(df[var],bins=20,label='original dataset')
    sns.distplot(New4[var],bins=20,label='cleaned by median dataset')
    plt.legend()
```

```python
plt.figure(figsize=(16,9))

for i, var in enumerate(new2):
    plt.subplot(2,2,i+1)
    
    sns.distplot(df[var],bins=20,label='original dataset')
    sns.distplot(New3[var],bins=20,label='cleaned by mean dataset')
    sns.distplot(New4[var],bins=20,label='cleaned by median dataset')
    plt.legend()
```

```python


for i, var in enumerate(new2):
    plt.figure(figsize=(16,16))
    plt.subplot(3,1,1)
    
    sns.boxplot(df[var],orient='h')
    
    plt.subplot(3,1,2)#mean
    sns.boxplot(New3[var],orient='h')
    
    plt.subplot(3,1,3)#median
    sns.boxplot(New4[var],orient='h')
  
```

```python
df_concat=pd.concat([df[new2],#original
                    New3[new2],#mean
                    New4[new2]],#median
                    axis=1)
```
